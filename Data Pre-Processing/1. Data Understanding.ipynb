```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# Dataset
heights = np.array([150, 160, 165, 170, 175, 180, 185, 190, 200])
```

```python
# Central Tendency
mean = np.mean(heights)
median = np.median(heights)
```

```python
# Calculate mode manually
unique, counts = np.unique(heights, return_counts=True)
max_count = np.max(counts)
if max_count > 1:
    mode = unique[counts == max_count]
else:
    mode = "No mode"

midrange = (np.min(heights) + np.max(heights)) / 2
```

```python
# Distribution Measures
Q1 = np.percentile(heights, 25)
Q3 = np.percentile(heights, 75)
IQR = Q3 - Q1
std_dev = np.std(heights)

```

```python
# Print results
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Midrange: {midrange}")
print(f"IQR: {IQR}")
print(f"Standard Deviation: {std_dev}")

# Statistical Graphs
# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(heights, kde=True)
plt.title('Histogram of Heights')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.show()

# Scatterplot
plt.figure(figsize=(10, 6))
plt.scatter(range(len(heights)), heights)
plt.title('Scatterplot of Heights')
plt.xlabel('Index')
plt.ylabel('Height (cm)')
plt.show()

# Quantile Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x=heights)
plt.title('Boxplot of Heights')
plt.xlabel('Height (cm)')
plt.show()

# QQ-Plot
plt.figure(figsize=(10, 6))
stats.probplot(heights, dist="norm", plot=plt)
plt.title('QQ-Plot of Heights')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.show()
```
